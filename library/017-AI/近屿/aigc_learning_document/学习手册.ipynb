# api接口

## api接口调用（1000次），使用完可以继续申请！申请方法联系小班或者添加下方客服！

## 每次调用gpt-3.5-turbo计一次次数，调用gpt-4计十次次数。失败不计次数。

## python调用示例

```python
# python调用示例
import requests

def generate_chat(api_key, model, user_message):
    url = "https://hr.aidexianzhaopinguan.com/cand/peixun/api/answer"
    
    payload = {
        "api_key": api_key,
        "model": model,
        "messages": [
            {"role": "system", "content": "你是聊天机器人，旨在和人类交流聊天"},
            {"role": "user", "content": user_message}
        ],
        "top_p": 1,
        "temperature": 0.9,
        "frequency_penalty": 0,
        "presence_penalty": 0
    }

    response = requests.post(url, json=payload)

    if response.status_code == 200:
        response_json = response.json()
        print(response_json)
        content = response_json['response']['choices'][0]['message']['content']
        return content
    else:
        return f"Error: {response.status_code} - {response.text}"

# 测试
api_key = "ojac-xxxxxxx"
model = "gpt-3.5-turbo"
user_message = "你好"

result = generate_chat(api_key, model, user_message)
print(result)
```
