## 一、开发环境的基本要求

### 1.1 硬件需求

**CPU/GPU配置**：高性能多核CPU对于并行处理任务至关重要，而高性能GPU（如NVIDIA的Tesla或Quadro系列）则是深度学习中不可或缺的，因为它们能显著加速模型的训练和推理过程。

    1. CPU：通用处理器，通常具有复杂的指令集。适用于一般性任务，包括操作系统管理、办公软件、编程等。
    
    2. GPU：具有简化的指令集，专注于并行计算任务。适用于图形渲染、科学计算、深度学习训练等需要大规模并行计算的任务。

**CPU的特性：**
- Low compute density: CPU通常拥有较少的算术逻辑单元（ALUs），这意味着它们在每个处理周期内可以执行的计算数量较少。
- Complex control logic: CPU设计用来处理复杂的任务和控制流，因此具有复杂的控制逻辑来管理指令的执行。
- Large caches (L1$/L2$, etc.): 为了快速访问数据，CPU有较大的缓存，如一级缓存（L1 Cache）和二级缓存（L2 Cache）。
- Optimized for serial operations: CPU是为连续执行单线程任务优化的，拥有更少的执行单元和高时钟速度来处理序列操作。
- Shallow pipelines (<30 stages): CPU的流水线阶段较少，这有助于快速完成任务，但每个任务处理的时间较长。
- Low Latency Tolerance: CPU对延迟的容忍度低，因为它们设计用来快速响应请求。
- Newer CPUs have more parallelism: 随着技术的发展，新一代的CPU增加了并行处理的能力，但仍然没有GPU那么多。

**GPU的特性：**
-  High compute density: GPU有很多并行执行单元（ALUs），因此它能在每个处理周期内执行大量的计算。
- High Computations per Memory Access: GPU在每次内存访问时可以执行更多的计算，这使得它们在处理大量数据时非常高效。
- Built for parallel operations: GPU是为并行操作设计的，尤其擅长图形处理，这是并行处理的典型案例。
- Deep pipelines (hundreds of stages): GPU有着深度流水线，可以同时处理多个阶段的计算任务。
- High Throughput: GPU设计用来提供高吞吐量，即每秒可以处理大量数据。
- High Latency Tolerance: GPU能够容忍较高的延迟，因为它们通过并行处理大量任务来隐藏延迟。
- Newer GPUs: 最新的GPU具有更好的流程控制逻辑，支持更复杂的数据访问模式，并且不再拥有单向流水线。

**内存要求**：至少16GB RAM，更大型的模型可能需要更多：

内存（RAM，Random Access Memory）是计算机中的一种主要硬件组件，用于存储临时数据。它是计算机的工作存储器，用于存储运行中的程序和正在处理的数据。

    1. 模型训练：
    训练大型模型通常需要大量的内存，特别是在处理大规模数据集时。内存越大，可以一次性加载更多的数据，减少数据加载的频率，提高训练效率。
    对于深度学习任务，模型的参数越多、层数越深，通常需要更多的内存来存储中间计算结果和梯度。
    
    2. 大型模型和数据集：
    大型的深度学习模型（如BERT、GPT等）和庞大的数据集需要更多的内存来进行高效的训练。
    当使用更大型的模型时，特别是在自然语言处理（NLP）和计算机视觉（CV）领域，16GB RAM 可能只能胜任一些中小型模型的训练。对于更大的模型，32GB、64GB甚至更多的内存可能是更好的选择。
    
    3. 并行性：并行计算时，每个计算单元都需要一部分内存。在多GPU或多CPU的情况下，内存需求可能会线性增加。
    使用多GPU时，还需要考虑模型和数据的分布，以确保每个GPU上的内存不会过载。
    
    4. 数据预处理：在数据预处理阶段，可能需要较大内存来处理原始数据、进行标准化、裁剪等操作。这对于在训练之前准备数据非常重要。
    
    5. 内存频率和带宽：内存的频率和带宽也是考虑因素。内存频率指的是内存模块的时钟速度，通常以MHz或GHz表示。它表示内存模块每秒钟可以进行多少次数据传输。更高频率的内存意味着它可以更快地执行读取和写入操作。内存带宽是指内存模块能够传输数据的最大速度，通常以GB/s表示。它是由内存模块的频率、总线宽度和双通道/四通道等因素决定的。更高的内存带宽可以提供更大的数据传输通道，从而在处理大量数据时更有效。


### 1.2 软件需求

#### **操作系统**：通常使用Linux，例如Ubuntu或CentOS。以下几个优势：

##### 开源性
- **源代码开放**：Linux的源代码是开放的，允许任何人查看、修改和分发。
- **与专有操作系统的对比**：不同于Windows和macOS的专有模式，Linux的源代码不是保密的。

##### 免费使用
- **通常免费提供**：大多数Linux发行版如Ubuntu、Fedora、Debian可以免费下载和使用。
- **企业级发行版**：一些企业级发行版可能会收取支持服务费用。

##### 多样性
- **不同的发行版**：Linux有多种不同版本，称为“发行版”（Distributions），例如Ubuntu、Fedora、Debian和CentOS。
- **各具特点**：这些发行版有着不同的界面、预装软件和用户体验。

##### 灵活性和可定制性
- **高度可定制**：用户可以更换不同的桌面环境、软件包和设置，以适应个人需求和偏好。

##### 稳定性和安全性
- **稳定和安全**：Linux被认为是非常稳定和安全的操作系统，广泛用于服务器和企业环境。
- **安全性特点**：其安全性得益于权限和用户管理系统，以及开源社区的持续支持和更新。

##### 社区支持
- **活跃的社区**：Linux拥有活跃且有帮助的用户和开发者社区。
- **快速问题解决**：作为开源系统，问题和漏洞通常可以通过社区迅速识别和修复。

##### 命令行界面
- **命令行的重要性**：尽管大多数Linux发行版提供图形用户界面（GUI），命令行界面对于系统管理、软件开发等任务仍然非常重要。

Linux广泛用于从个人电脑到服务器、嵌入式系统乃至超级计算机等各种环境，其灵活性和开源特性使其成为众多领域的首选操作系统。

## 二、搭建步骤与配置
安装操作系统和必要的软件包，安装编译器和工具链

### 2.1 操作系统和必要的软件包：
确保操作系统是最新的，安装必要的编译器和工具链。

#### 首先更新操作系统以确保所有软件包是最新的。
    Ubutu:
    sudo apt update        # 更新软件包列表
   
    sudo apt upgrade       # 升级所有可升级的软件包

    centOs:
    sudo yum update        # 更新软件包列表
    
    sudo yum upgrade       # 升级所有可升级的软件包


| 功能                | `apt`                                       | `yum`                                       |
|---------------------|---------------------------------------------|---------------------------------------------|
| **包管理系统**       | 使用 Debian 软件包管理系统（dpkg）              | 使用 Red Hat 软件包管理系统（RPM）              |
| **发行版支持**       | 主要用于 Debian 系列发行版，如 Ubuntu           | 主要用于 Red Hat 系列发行版，如 CentOS 和 Fedora |
| **命令语法**         | 使用 `apt` 命令，如 `apt update` 和 `apt upgrade` | 使用 `yum` 命令，如 `yum update` 和 `yum upgrade` |
| **软件源配置**       | 使用 `/etc/apt/sources.list` 文件来配置软件源    | 使用 `/etc/yum.conf` 和 `/etc/yum.repos.d/` 目录下的配置文件来配置软件源 |
| **依赖解决**         | 使用 Advanced Packaging Tool（APT）来处理软件包的依赖关系 | 使用 Yellowdog Updater, Modified（YUM）来处理软件包的依赖关系 |


#### 安装编译器和工具链

    对于大多数软件开发项目，GCC/G++编译器、Make等工具是必需的。
   
    sudo apt install build-essential    # 安装编译器和相关工具

*build-essential： 是一个元包（meta-package），它实际上并不包含可执行的程序，而是依赖于一组用于软件开发的基本工具和库。安装 build-essential 将安装一组通常在编译软件时所需的工具，包括编译器（如 GCC）、构建工具（如 make）、标准 C 库头文件等。

**编译器（GCC）：**

- **作用：** 编译器负责将高级编程语言（如C、C++等）编写的源代码翻译成机器语言或汇编语言，生成可执行文件。
- **特点：** GCC 是一个开源的编译器套件，支持多种编程语言，并可在多个平台上运行。它包含预处理器、编译器、汇编器和链接器等组件。

**构建工具（make）：**

- **作用：** 构建工具负责自动化软件项目的构建过程，包括编译、链接和其他必要的步骤。make 工具通过读取一个称为 Makefile 的文件，其中包含了项目的构建规则和依赖关系，然后根据这些规则执行相应的操作。
- **特点：** make 工具提高了软件开发的效率，尤其对于大型项目，可以确保只有发生变化的源文件才会重新编译，从而节省时间。

**标准 C 库头文件：**

- **作用：** 标准 C 库头文件包含了在C语言中常用的函数和宏的声明，如输入输出函数、字符串处理函数等。它提供了许多基本的功能，开发者可以直接使用这些功能，而无需从零开始编写。
- **特点：** 标准 C 库头文件是为了提供跨平台的可移植性而设计的，开发者可以在不同的系统上使用相同的头文件，而无需担心特定系统的差异。

**Vim 安装指南**

- Windows

1. **下载 Vim 安装程序**
   - 访问 [Vim 官网](https://www.vim.org/download.php)。
   - 在 "PC: MS-DOS and MS-Windows" 部分，选择 "Self-installing executable" 链接下载 `.exe` 安装程序。

2. **安装 Vim**
   - 双击下载的 `.exe` 文件以启动安装程序。
   - 按照提示完成安装。

3. **验证安装**
   - 打开命令提示符（CMD）或 PowerShell，输入 `vim` 并回车。如果 Vim 启动，说明安装成功。

- macOS

1. **使用 Homebrew 安装**
   - 打开终端。
   - 如果尚未安装 Homebrew，请首先安装 Homebrew（访问 [Homebrew 官网](https://brew.sh/) 获取安装命令）。
   - 运行以下命令安装 Vim：
     ```bash
     brew install vim
     ```

2. **验证安装**
   - 在终端中输入 `vim` 并回车。如果 Vim 启动，说明安装成功。

- Linux

1. **使用包管理器安装**
   - **Debian/Ubuntu** 系统：
     ```bash
     sudo apt update
     sudo apt install vim
     ```

   - **Fedora** 系统：
     ```bash
     sudo dnf install vim
     ```

   - **Arch Linux** 系统：
     ```bash
     sudo pacman -S vim
     ```

2. **验证安装**
   - 打开终端，输入 `vim` 并回车。如果 Vim 启动，说明安装成功。



#### vim 使用基本方法

1. 启动Vim：打开终端，并输入以下命令启动Vim编辑器：vim

2. 基本的模式切换，Vim有两种主要的模式：命令模式和插入模式。

- 命令模式：默认启动后的模式，用于导航和执行命令。
- 插入模式：按下 i 键进入插入模式，可以输入文本。

3. 基本导航 在命令模式下进行文本导航：使用箭头键或 h, j, k, l 进行上下左右导航。
- w：向前移动一个单词。
- b：向后移动一个单词。
- 0：移动到行首。
- $：移动到行尾。
- gg：跳转到文件开头。
- G：跳转到文件末尾。

4. 文本编辑 在命令模式下进行文本编辑：

- x：删除当前光标所在处的字符。
- dd：删除当前行。
- yy：复制当前行。
- p：粘贴文本。

5. 保存和退出 在命令模式下保存和退出：

- :w：保存文件。
- :q：退出Vim。
- :wq：保存并退出。
- :q!：强制退出（不保存）。

6. 其他常用命令
- :set number：显示行号。
- :set nonumber：隐藏行号。
- :find filename：查找并打开名为 filename 的文件。
- :help：获取帮助文档。

**配置Python虚拟环境**

使用Anaconda或Miniconda创建隔离的Python环境，这有助于管理不同项目的依赖。Anaconda 和 Miniconda都是流行的开源分发版，用于管理 Python的环境和包。它们都由 Anaconda, Inc. 维护

**Anaconda:**

- **全功能分发版：** Anaconda 是一个包含了 Python 语言、科学计算和数据科学所需的数百个第三方库的大型分发版。

- **适合初学者：** 它包含了许多预安装的包，这对于初学者来说很方便，因为他们不需要单独安装这些包。

- **附加工具：** Anaconda 还包括 Anaconda Navigator，一个图形用户界面，可以方便地管理环境和包，以及启动应用程序和工具。

- **体积较大：** 由于预装了许多包，Anaconda 的下载体积相对较大。


**Miniconda:**

- **最小化安装：** Miniconda 提供了一个最小化的安装器，只包含了 Conda 和它的依赖，以及 Python。

- **灵活性更高：** 用户可以使用 Conda 创建自己所需的环境，并仅安装所需的包。

- **更适合有经验的用户：** 对于希望完全控制其安装环境的用户来说，Miniconda 是一个更好的选择。

- **占用空间更小：** 由于不预装许多包，Miniconda 的下载和安装后所占用的空间更小。


**Conda vs Pip:**

| 特征                 | Conda                                           | Pip                                              |
|----------------------|-------------------------------------------------|---------------------------------------------------|
| **包管理系统**        | Conda 是一个综合的包管理系统和环境管理系统。      | Pip 是Python的包管理系统，主要用于安装Python包。     |
| **环境管理**          | Conda 可以创建和管理虚拟环境，不仅限于Python。    | Pip 通常与虚拟环境工具（如conda）一起使用。      |
| **语言支持**          | Conda 可以管理任何语言的包，不仅仅是Python。       | Pip 主要用于Python包的管理。                          |
| **依赖解决**          | Conda 可以处理包的依赖关系，确保环境中的所有包都兼容。| Pip 依赖解决有时可能会出现问题，需要手动处理。         |
| **源**               | Conda 使用Anaconda或Miniconda仓库，默认包含众多科学计算包。| Pip 使用Python Package Index (PyPI) 作为默认源。      |
| **速度**              | Conda 安装速度较快，因为它可以预先编译二进制文件。  | Pip 在安装过程中通常需要编译源代码，可能较慢。         |
| **跨平台性**         | Conda 可以在Windows、Linux和macOS上无缝运行。       | Pip 同样跨平台，但在某些特定情况下可能需要手动配置。   |
| **可执行文件**       | Conda 安装的包中可能包含可执行文件，可用于任何环境。 | Pip 安装的包通常只提供Python代码，可执行文件需要手动添加到系统路径。 |

这些是Conda和Pip的主要区别和联系。通常情况下，Conda更适合处理复杂的环境和跨语言的项目，而Pip则更适合Python特定的包管理。在一些项目中，两者可以结合使用，以充分发挥各自的优势。


**注意**：避免混合使用 Conda 和 Pip 安装相同的包，以免导致依赖冲突。

conda --version

conda create --name env_name package_name

conda activate env_name

#### 退出环境:

conda deactivate

#### 列出所有已创建的环境:

conda env list 或者 conda info --envs

conda install --name env_name package_name

例如，在名为“myenv”的环境中安装SciPy：conda install --name myenv scipy

#### 更新包:

conda update package_name

例如，更新NumPy包：conda update numpy

#### 更新所有包在当前环境:

conda update --all

#### 删除包:

conda remove package_name

例如，删除NumPy包：conda remove numpy


#### 删除环境:

conda remove --name env_name --all

例如，删除名为“myenv”的环境：conda remove --name myenv --all

#### 查找包:

conda search package_name

例如，搜索NumPy包：conda search numpy

#### 列出环境中安装的包:

conda list

在特定环境中列出包：conda list -n env_name


#### 导出环境:

conda env export > environment.yml

导出当前环境的配置到YAML文件。

这个 environment.yml 文件可以用于在其他地方重新创建相同的环境。例如，如果您要在另一台计算机上或共享给其他人时，他们只需运行以下命令来创建相同的环境：

### 2.8 调试工具
Python的pdb：Python的内置交互式调试器，可以用来设置断点、逐步执行代码、检查变量等。

在IDE中使用调试工具，如PyCharm或VSCode的调试功能。

#### Pdb

Python的内置交互式调试器是pdb（Python Debugger）。使用pdb，你可以设置断点、逐步执行代码、检查变量等，以帮助调试和理解代码的执行过程。

- **定位和解决错误：** PDB 是一种交互式调试工具，它允许你在程序执行的特定点停下来，检查变量的值，逐步执行代码，并发现潜在的错误。这对于调试程序中的 bug 或理解代码的执行流程非常有帮助。

- **逐步执行代码：** PDB 允许你逐步执行代码，一步一步地查看程序在每个步骤中的状态。这有助于理解代码的运行逻辑，特别是对于复杂的或不熟悉的代码。

- **检查变量：** 在 PDB 中，你可以在程序暂停时检查变量的值，确保它们与你的预期一致。这对于追踪变量的值如何在代码中变化以及找到问题所在非常有帮助。

- **条件断点：** PDB 允许你设置条件断点，即只有当某个条件满足时，程序才会在特定点暂停。这对于在特定情况下调试代码非常有用。

- **交互式调试：** PDB 提供了一个交互式的命令行界面，让你可以在程序执行期间与调试器进行交互。你可以执行各种命令，例如查看变量、设置断点、执行代码片段等。

- **不需要额外工具：** PDB 是 Python 的标准库的一部分，因此不需要额外的安装或配置。你可以直接在终端或脚本中使用 PDB 进行调试。

#### **Python的pdb调试步骤**

**导入pdb**
在Python脚本中，首先需要导入`pdb`模块。

**设置断点**
你可以在代码中的任何位置添加`pdb.set_trace()`来设置一个断点。程序执行到这一行时会暂停，并进入调试模式。

**运行程序**
运行程序时，一旦执行到达断点，程序将会暂停。这时，你会看到一个交互式的调试提示符。

**调试命令**
在调试模式下，可以使用多种命令来检查程序的当前状态。一些常用的命令包括：
- `l` (list): 显示当前行的上下文代码。
- `n` (next): 执行下一行代码。
- `c` (continue): 继续执行程序，直到遇到下一个断点。
- `p` (print): 打印一个变量的值。
- `q` (quit): 退出调试器。

**调试流程**
使用这些命令，可以逐步执行代码，检查变量的值，评估表达式等，这对于找出代码中的错误和理解代码的执行流程非常有帮助。


````python
# 导入Python的调试模块pdb
import pdb

# 定义一个函数sum_numbers，用于计算一个数字列表的总和
def sum_numbers(num_list):
    total = 0  # 初始化总和变量total为0
    for num in num_list:  # 遍历传入的数字列表num_list
        total += num  # 将当前数字加到总和total上
        pdb.set_trace()  # 在循环中的每一次迭代后设置一个断点
    return total  # 当遍历完整个列表后，返回计算的总和

# 创建一个数字列表
numbers = [1, 2, 3, 4, 5]
# 调用sum_numbers函数，并将numbers列表作为参数传入，计算列表中所有数字的和
result = sum_numbers(numbers)
# 打印计算结果
print("结果是:", result)
````

4.基于华佗数据集处理成百川2数据集的代码示例：

```python
import json

input_file = 'format_data.jsonl'
output_file = 'baichuan_formatted_data.json'
output_file_test = 'baichuan_formatted_data_test.json'

baichuan_data = []
baichuan_data_test = []
count = 0
# 处理每行数据并写入输出文件
with open(input_file, 'r') as f_in:
    for line in f_in:
        count+=1
        data = json.loads(line)
        
        # 提取华佗数据集中的问题和回答
        human_value = data["question"]
        assistant_value = data["answer"]
        _id = data["id"]
        
        # 构建百川数据格式
        # baichuan_data = {
        #     "instruction": f"Human:{human_value}\nAssistant:",
        #     "input": '',
        #     "output": assistant_value
        # }
        _data = {"id": _id,"conversations": [{"from": "human", "value": f"{human_value}\n"}, {"from": "gpt","value": f"{assistant_value}\n"}]}
        if count  <= 5000:
            # 前5000条做训练集
            baichuan_data.append(_data)
        elif count <= 6000:
            # 5000-6000条做测试集
            baichuan_data_test.append(_data)
        else:
            break
        
        
with open(output_file, 'w', encoding='utf-8') as f_out:
    # 将处理后的数据写入输出文件
    json.dump(baichuan_data, f_out, ensure_ascii=False)

with open(output_file_test, 'w', encoding='utf-8') as f_out:
    # 将处理后的数据写入输出文件
    json.dump(baichuan_data_test, f_out, ensure_ascii=False)
```
