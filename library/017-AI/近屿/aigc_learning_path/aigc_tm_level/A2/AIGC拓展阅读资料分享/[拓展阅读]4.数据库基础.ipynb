# 数据库基础

数据库，简而言之就是存储数据的库。基本操作就是增删改查（增加，删除，修改和查询），数据库有很多种，常见的数据库有mysql（新手必学，开源免费），oracle，sql server，mongodb等。
我们在运行程序时，有变量，为什么还需要数据库来存储数据呢，我们平时程序里的变量都存在内存里，下次再重新运行的时候数据就消失了，而且就算存储在文件中，也会出现数据管理的问题（什么格式，如何访问），况且还有多用户访问问题。
接下来我们以一些典型的数据库为例来介绍数据库的基础内容和操作方式。

- 重命名表

- 修改表结构

  
ALTER TABLE 表名 ADD COLUMN 列名 数据类型;

ALTER TABLE 表名 DROP COLUMN 列名;

ALTER TABLE 表名 MODIFY COLUMN 列名 新数据类型;


### 2、mysql介绍及使用
mysql属于关系型数据库，看起来就像是excel的表格那样。

<img src="images/20.png" style="margin-left: 0px" width="1600px">


如图数据的组织结构分为数据库（database）、数据表（table）、数据行（row）、字段（field）这四部分组成。图中所示是一个bookstore0612的数据库中有一个叫做book的表，其中展示了7条数据，每条数据有一些字段。
除了这些东西之外还有索引，视图等概念。
索引是用来快速定位数据和访问数据的数据结构，这涉及到数据库的底层存储方式，感兴趣的学员可以自己搜索一下：mysql索引原理。
视图是一个或者多个表的查询结果集，可以简化复杂的数据查询和提供数据安全性，就比如上面这张图，我们只想看见id和title字段，那我们就可以建立对应的视图。

**sql语句**

sql语句就是操作数据库的语句，有些类似于编程代码，不同数据库对应的sql语句大同小异，我们这里以mysql的为例。
我们以上面这张图为例，执行增删改查操作。

- 查询所有书籍信息

SELECT * FROM books;

- 查询特定列（书名和作者）

SELECT title, author FROM books;

- 查询价格低于30元的书籍
  
SELECT * FROM books WHERE price < 30;

- 查询作者为东野圭吾的书籍
  
SELECT * FROM books WHERE author = '东野圭吾';

- 按价格降序排列书籍
  
SELECT * FROM books ORDER BY price DESC;

- 添加新书籍
  
INSERT INTO books (title, author, price, sales, stock, img_path)
VALUES ('边城', '沈从文', 25.00, 50, 80, 'static/img/');

- 更新书籍价格（将《解忧杂货店》价格调整为30元）
  
UPDATE books SET price = 30.00 WHERE title = '解忧杂货店';

- 更新库存（减少《百年孤独》库存5本）

UPDATE books SET stock = stock - 5 WHERE title = '百年孤独';

- 删除特定书籍（删除ID为2的记录）

DELETE FROM books WHERE id = 2;

- 计算所有书籍的平均价格

SELECT AVG(price) AS average_price FROM books;

- 按作者分组统计书籍数量和总销量

SELECT author, COUNT(*) AS book_count, SUM(sales) AS total_sales
FROM books GROUP BY author;

- 查询库存少于100本的书籍

SELECT title, stock FROM books WHERE stock < 100;

- 查询最畅销的前3本书

SELECT title, sales FROM books ORDER BY sales DESC LIMIT 3;

- 开始事务（对应界面中的"开始事务"按钮）

START TRANSACTION;

- 在事务中执行多个操作

UPDATE books SET stock = stock - 1 WHERE id = 1;
INSERT INTO cart_items (book_id, quantity) VALUES (1, 1);

- 提交事务

COMMIT;

- 或回滚事务

ROLLBACK;

所谓事务，是进行数据库操作的基本单位，可以是一组数据库操作，为了保证操作前后的数据一致性，这些操作要么全部成功，要么全部失败。
事务有如下特征：

- 1、原子性(Atomicity)：事务中的所有操作要么全部完成，要么全部不完成，不会停留在中间状态。
- 2、一致性(Consistency)：事务执行前后，数据库从一个一致状态转变为另一个一致状态。
- 3、隔离性(Isolation)：多个事务并发执行时，一个事务的执行不应影响其他事务。
- 4、持久性(Durability)：事务一旦提交，其结果就是永久性的，即使系统故障也不会丢失。

## 二、非关系型数据库

非关系型数据库（NoSQL）是为解决大规模数据存储和高并发访问而设计的新型数据库系统，与传统的SQL关系型数据库有着显著区别。

Nosql数据库的核心特点

- 1、非关系型数据模型：不适用固定的表结构
- 2、水平扩展能力：易于分布式部署
- 3、高性能：针对特定场景优化
- 4、灵活的数据结构：支持半结构化/非结构化数据
- 5、高可用性：自动故障转移和复制
  
主要的NoSQL数据库类型：


#### 1、键值数据库（key-value stores）
  
代表产品：redis、DynamoDB、Riak

特点：
- 最简单的NoSql模型；
- 数据以键值形式存储；
- 极高的读写性能；
- 常用于缓存和会话存储；

<img src="images/21.png" style="margin-left: 0px" width="1600px">


#### 2、文档数据库
 
代表产品：MongoDB、CouchDB、Firestore
 
特点：
 
- 数据存储为文档（json/bson)
- 文档可包含嵌套结构
- 灵活的查询能力
- 常用于内容管理系统、用户配置文件

<img src="images/22.png" style="margin-left: 0px" width="1600px">


#### 3、列族数据库（Column-Family Stores）

代表产品：Cassandra、HBase、ScyllaDB

特点：
- 数据按照列族组织
- 适合超大规模数据集
- 高写入吞吐量
- 常用于时间序列数据、日志分析

<img src="images/23.png" style="margin-left: 0px" width="1600px">


#### 4、图数据库（Graph Databases）

代表产品：Neo4j、ArangoDB、JanusGraph

特点：
- 以节点、边和属性存储数据
- 高效处理复杂关系
- 专门的关系查询语言
- 常用于社交网络、推荐系统、欺诈检测

### NoSQL适用场景：


- 1.大数据应用​：处理PB级数据
- 2.实时分析​：快速处理流数据
- 3.内容管理系统​：灵活的内容结构
- 4.物联网应用​：处理设备生成的海量数据
- 5.个性化推荐​：处理复杂用户关系
- 6.快速迭代开发​：无需预先定义严格模式
