SWE-bench的实例中，会提供一段问题的描述，以及相应的基础代码上下文。这时候我们会发现，虽然问题描述往往比较短（平均195个英文单词），但相应的基础代码上下文却会长得多（平均43万行代码），要把这么长的文本直接输给大模型，往往会超过大模型最大上下文窗口的限制。所以，我们需要用特定的检索方法，来选取与问题最相关的内容，输入给模型。
为了解决这种问题，这篇文章作者简单地使用一个通用检索系统来选择要插入的文件作为上下文。这里有两种方法：
- 1)稀疏检索（使用BM25算法）
- 2)oracle检索。

“稀疏检索”的设置使用了BM25算法。BM25（Best Matching 25）是一种经典的信息检索算法。其被广泛应用于信息检索领域的排名函数，用于估计文档D与用户查询Q之间的相关性。它是一种基于概率检索框架的改进，特别是在处理长文档和短查询时表现出色。BM25的核心思想是基于词频(TF)和逆文档频率(IDF)来,同时还引入了文档的长度信息来计算文档D和查询Q之间的相关性。所以在给大模型提出问题A的同时，使用BM25算法找到最相关的文件B，将其和问题A一起输入大模型。
“oracle检索”的设置其实有点像“作弊”，因为这种方法在现实实际使用中是比较难实现的。这种方法直接选取那些在Github上的回复issue中修改了的那些文件，比如有人为了解决问题A，修改了文件B，那么他会在给大模型提出问题A的同时直接把文件B输给大模型。然而现实中如果一个工程师想要解决某个问题，他是不能直接知道问题出自哪个文件的。不过作为一种评测方式，这倒是合理的。

````python
#写出一个正则表达式，能够提取字符串的所有整数和小数

# 导入正则表达式模块
import re
# 定义函数find_num，用于查找字符串中的数字
def find_num(str):
    # 使用正则表达式查找字符串中的数字
    num = re.findall(r'[-+]?[0-9]+\.?[0-9]*',str)
    # 返回查找到的数字
    return num
# 定义字符串str
str = "12.33今天是星期一3"
# 调用函数find_num，查找字符串中的数字
num = (find_num(str))
# 打印查找到的数字
print(num)
````

比如现在我们要让大模型写出一个正则表达式，能提取字符串的所有整数和小数。可以看到它自动import了re模块，然后定义了一个叫find_num的函数，函数接受一个字符串作为参数。这说明，Copilot很好地理解了我们注释提出的要求。在使用具体提取字符串函数的时候，Copilot成功使用了需要的re.findall函数，用来找到str中所有符合前面正则表达式的子串。可以看到，正则表达式是正确的。
正则表达式 `r'\d+\.?\d*'` 是用来匹配整数或浮点数的。这个正则表达式的含义是：
- `\d+`：匹配一个或多个数字
- `\.?`：匹配零个或一个小数点
- `\d*`：匹配零个或多个数字
这个正则表达式可以匹配如 "123"、"123."、"123.456" 这样的字符串。

上图演示中，Copilot简单生成一个正则表达式用来匹配字符串里面的所有数字。这些简单的函数Copilot实现得相当好，而且能够让程序员更专注于主题而不是每个小函数的细枝末节。

````python
#打印路径下所有txt文件的名字

# 导入os模块
import os
# 定义一个函数，用于查找指定路径下的txt文件
def find_txt(path):
    # 遍历指定路径下的所有文件
    for root,dirs,files in os.walk(path):
        # 遍历文件
        for file in files:
            # 判断文件后缀是否为txt
            if os.path.splitext(file)[1] == '.txt':
                # 打印文件名
                print(file)
# 获取当前路径
path = os.getcwd()
# 调用函数，查找指定路径下的txt文件
find_txt(path)
````

上图演示中，copilot写一个小程序，能够直接找出当前文件夹下所有txt文件的名字。有趣的是，这次注释里仅仅说了“打印路径下所有txt文件的名字”，相关的整段代码一下子就生成出来了。os.walk()用来找出路径下的各种文件夹和文件，然后选择找到的文件列表中，以“.txt”为后缀的所有文件，把他们的名字打印出来。

````python
# 写一个函数，能够判断一个无序整数数组中第K大的数是多少
def find_Kth_largest(nums, k):
    # 使用快速排序的 partition 函数，找到第 k 大的数
    left, right = 0, len(nums) - 1
    while left <= right:
        pivot_index = partition(nums, left, right)
        if pivot_index == k - 1:
            return nums[pivot_index]
        elif pivot_index > k - 1:
            right = pivot_index - 1
        else:
            left = pivot_index + 1
            
    return None


def partition(nums, left, right):
    pivot = nums[left]
    while left < right:
        while left < right and nums[right] >= pivot:
            right -= 1
            
        nums[left] = nums[right]
        
        while left < right and nums[left] <= pivot:
            left += 1
            
        nums[right] = nums[left]
        
    nums[left] = pivot
    return left

# 示例
nums = [3, 2, 1, 55, 62, 42]
k = 2
print(find_Kth_largest(nums, k))  # 输出第2大的数，即55
````

````python
import pygame
import random

# 初始化pygame
pygame.init()

# 定义屏幕大小和标题

# 设置屏幕宽高
screen_width = 960
screen_height = 960
# 设置屏幕大小
screen = pygame.display.set_mode((screen_width, screen_height))
# 设置标题
pygame.display.set_caption("飞机大战")

# 定义颜色
WHITE = (255, 255, 255)

# 定义飞机类

# 定义一个Plane类，继承自pygame.sprite.Sprite类
class Plane(pygame.sprite.Sprite):
   # 初始化函数，用来初始化Plane类的属性
   def __init__(self):
       # 调用父类的初始化函数，用来初始化父类的属性
       super().__init__()
       # 创建一个50x50的Surface对象
       self.image = pygame.Surface((50, 50))
       # 填充Surface对象，颜色为白色
       self.image.fill(WHITE)
       # 获取Surface对象的矩形对象
       self.rect = self.image.get_rect()
       # 设置矩形的左上角坐标，使其居中显示
       self.rect.topleft = (screen_width / 2 - self.rect.width / 2, screen_height / 2 - self.rect.height / 2)

   # 定义update函数，用于更新游戏中的角色位置
def update(self):
       # 获取按键状态
       keys = pygame.key.get_pressed()
       # 如果按下左键且角色位置不超出屏幕最左端，则将角色向左移动5个像素
       if keys[pygame.K_LEFT] and self.rect.left > 0:
           self.rect.x -= 5
       # 如果按下右键且角色位置不超出屏幕最右端，则将角色向右移动5个像素
       if keys[pygame.K_RIGHT] and self.rect.right < screen_width:
           self.rect.x += 5
       # 如果按下上键且角色位置不为0，即不超出屏幕顶端，则将角色向上移动5个像素
       if keys[pygame.K_UP] and self.rect.top > 0:
           self.rect.y -= 5
       # 如果按下下键且角色位置不超出屏幕最底端，则将角色向下移动5个像素
       if keys[pygame.K_DOWN] and self.rect.bottom < screen_height:
           self.rect.y += 5

# 定义子弹类
           
# 定义一个Bullet类，继承自pygame.sprite.Sprite类
class Bullet(pygame.sprite.Sprite):
   # 初始化函数，传入参数plane
   def __init__(self, plane):
       # 调用父类的初始化函数
       super().__init__()
       # 创建一个10x10的矩形，填充白色
       self.image = pygame.Surface((10, 10))
       self.image.fill(WHITE)
       # 获取矩形的矩形框
       self.rect = self.image.get_rect()
       # 设置矩形的中心点为plane的矩形框的中心点，上边距为plane的矩形框的上边距
       self.rect.midtop = (plane.rect.centerx, plane.rect.top)
       # 设置速度为5
       self.speed = 5

   # 更新函数，用于更新矩形框的位置
   def update(self):
       # 矩形框的y坐标减去速度
       self.rect.y -= self.speed

# 创建飞机精灵组
planes = pygame.sprite.Group()
# 创建一个空的分组
plane = Plane()
# 创建一个飞机对象
planes.add(plane)

# 创建子弹精灵组
bullets = pygame.sprite.Group()

# 游戏主循环
running = True
while running:
   # 填充背景色
   screen.fill((0, 0, 0))
   # 将飞机绘制到屏幕上
   screen.blit(plane.image, plane.rect.topleft)

   # 处理事件
   for event in pygame.event.get():
       # 如果用户点击关闭按钮，则退出游戏
       if event.type == pygame.QUIT:
           running = False
       elif event.type == pygame.KEYDOWN:
           # 如果按下空格键，则创建子弹
           if event.key == pygame.K_SPACE:
               bullet = Bullet(plane)
               bullets.add(bullet)

   # 更新飞机位置
   planes.update()

   # 更新子弹位置
   bullets.update()

   # 检测碰撞
   collided_bullets = pygame.sprite.groupcollide(bullets, planes, True, True)

   # 重新创建子弹
  # 遍历碰撞的子弹，并创建一个新的子弹
   for collided_bullet in collided_bullets:
       # 创建一个新的子弹
       bullet = Bullet(plane)
       # 将新的子弹添加到子弹列表中
       bullets.add(bullet)

   # 绘制子弹
   bullets.draw(screen)

   # 更新屏幕
   pygame.display.update()

# 退出pygame
pygame.quit()
````

我们用pass@k这一度量标准来评估代码生成模型的功能正确性。该标准的核心思想是，对于每个编程问题，生成k个不同的代码样本，只要其中任何一个样本能够通过单元测试，就认为该问题已被解决。研究者们会统计所有问题中成功解决的比例，以此来衡量模型的性能。aiXcoder 7B在评测的时候使用了Pass@1指标，即对每个评测集的每个问题都生成一个答案，看这一个答案是否正确，最后准确率为正确的问题/评测集所有问题。

Decoder-only架构通常用于自回归模型，其中最著名的例子是OpenAI的GPT（Generative Pre-trained Transformer）系列模型。这种架构完全由解码器组成，没有独立的编码器。在训练过程中，模型接收一个序列的前缀作为输入，然后尝试预测序列的下一个元素，直到序列的末尾。

优点：

    训练效率： 因为不需要编码器和解码器之间的交互，训练过程较为简单和高效。
    泛化能力： Decoder-only模型，尤其是自回归模型，能够学习到文本中的长距离依赖关系，提升在多种任务上的泛化能力。
    语言生成： 这种架构非常适合语言生成任务，因为它可以逐字生成文本，每次生成一个词，且生成的文本依赖于先前生成的词。

缺点：

    并行性： 在生成文本时，由于需要等待前一个词的生成结果才能生成下一个词，这限制了并行计算的能力。
    效率问题： 对于长序列的生成，逐字生成可能会导致较长的生成时间和较高的计算成本。

Encoder-only架构，如Google的BERT（Bidirectional Encoder Representations from Transformers），主要关注于生成高质量的文本表示。这类模型通常使用双向的Transformer编码器来处理输入文本，从而得到一个包含文本全局信息的向量表示。

优点：

    双向理解： 编码器可以同时考虑文本的前向和后向上下文，这有助于模型理解词语的多义性和语境。
    预训练与微调： 这类模型通常通过大规模无监督预训练学习通用的语言表示，之后可以在特定任务上进行微调，以达到很好的性能。

缺点：

    生成能力有限： Encoder-only架构本身不具备生成序列的能力，通常需要额外的解码器或策略来实现生成任务。
    序列理解： 尽管在理解单个句子方面表现出色，但在处理长序列或跨句的依赖关系时，其表现可能不如Decoder-only或Encoder-Decoder架构。

### 4.1 生成算法框架

#### 使用大模型生成目标任务的算法框架，你可以按照以下步骤操作：

1. 明确你的目标任务：你需要解决什么问题？你希望达到什么目标？

2. 描述你的需求：向模型描述你的需求，包括你希望解决的问题、问题的条件和预期的结果。

3. 生成算法框架：请求模型生成算法框架。你可以使用类似于“为这个问题生成一个算法框架”这样的指令。

4. 检查生成的算法框架：检查模型生成的算法框架是否符合你的需求，是否能够有效地解决你的问题。

以下是一个Python程序的例子：

````python
# 你的问题是：如何实现一个排序算法？
# 你可以向模型请求：“为这个问题生成一个算法框架”
# 模型可能会生成以下的算法框架：

def sort_algorithm(arr):
    # 第一步： 检查输入是否合法
    if len(arr) <= 1:
        return arr

    # 第二步：选择一个基准值
    pivot = arr[len(arr) // 2]

    # 第三步：将数组分为三个子数组，分别包含小于、等于、大于基准值的元素
    less = [x for x in arr if x < pivot]
    equal = [x for x in arr if x == pivot]
    greater = [x for x in arr if x > pivot]

    # 第四步：递归地对小于和大于基准值的子数组进行排序
    return sort_algorithm(less) + equal + sort_algorithm(greater)

array=[3,1,5,7,2]
print(sort_algorithm(array))
````

通过对任务需求的清晰认知，以及大模型对细节更好的把控，程序员可以更高效地解决复杂问题。但是一定要记住，生成好的算法框架前提是，程序员已经对整个任务有一个清晰的结构认知，才能不断拆解任务，让大模型去生成各个组件！

### 5.2. 使用编程大模型生成代码文档

文档相当重要！然而，很多程序员在维护复杂项目的时候，不知道文档从何下手。其实编程大模型在生成代码文档方面对程序员的帮助也能相当大。

使用编程大模型生成代码文档的步骤可以如下：

1. 首先，明确你希望生成的文档的类型和内容。这可能是函数的注释、模块的说明、或者整个项目的文档。

2. 然后，将你的需求以自然语言的形式输入到模型中。例如，你可以输入“为这个函数生成一个注释”，然后提供函数的代码。

3. 模型将根据你的输入生成相应的文档。你可能需要检查并修改模型的输出，以确保它准确地描述了代码的功能和行为。

4. 最后，将生成的文档添加到你的代码中。

Copilot甚至直接集成了生成文档的功能。就以上面的排序代码为例，假设注释都被删除：

````python
def sort_algorithm(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    less = [x for x in arr if x < pivot]
    equal = [x for x in arr if x == pivot]
    greater = [x for x in arr if x > pivot]
    return sort_algorithm(less) + equal + sort_algorithm(greater)
````

````python
# 简单生成样例：
def sort_algorithm(arr):
    """
    这是一个使用快速排序算法的函数。

    快速排序是一种分而治之的排序算法，它将一个数组分为两个子数组，然后递归地对它们进行排序。

    参数:
    arr: 需要排序的列表。

    返回:
    返回排序后的列表。

    示例:
    >>> sort_algorithm([3, 6, 8, 10, 1, 2, 1])
    [1, 1, 2, 3, 6, 8, 10]
    """
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    less = [x for x in arr if x < pivot]
    equal = [x for x in arr if x == pivot]
    greater = [x for x in arr if x > pivot]
    return sort_algorithm(less) + equal + sort_algorithm(greater)
````

### 5.3. 安全性问题

AI大模型在生成代码时，可能会存在以下几种安全性问题：

AI大模型生成代码可能存在以下几个安全性问题：

1. 信息泄露：如果模型在训练过程中接触到敏感信息，可能会在生成代码时无意中泄露这些信息。

2. 不安全的代码：AI可能生成具有安全漏洞的代码，例如未经验证的用户输入、不正确的权限管理等。

3. 不可预测的行为：AI生成的代码可能会有不可预测的行为，可能会导致系统崩溃或者其他未预期的结果。

4. 依赖问题：AI可能会生成依赖于特定库或者版本的代码，这可能会导致安全问题，特别是当这些库或者版本存在已知的安全漏洞时。

找到可能的问题后，我们也可以用前面学过的方法，在写代码的时候就用编程大模型代码补全和注释生成代码的能力，直接得到想要的结果。下面是一个例子，我们在写代码时直接提示AI在代码补全时使用哈希函数来存储和验证密码，以提高安全性：

````python
import hashlib
from flask import Flask, render_template, request, redirect

# 创建一个Flask实例
app = Flask(__name__)

# 储存哈希过的密码，而不是明文密码
hashed_password = hashlib.sha256('123456'.encode()).hexdigest()

# 添加一个路由
@app.route('/')
def hello_world():
    if request.method == 'GET':
        return redirect('login')
    
@app.route('/login', methods=['GET', 'POST'])
def login():
    if request.method == 'GET':
        print('666')
        return render_template('login2.html')
    elif request.method == 'POST':
        username = request.form.get('username')
        password = request.form.get('password')
        # 使用哈希函数验证密码！！！
        if username == 'admin' and hashlib.sha256(password.encode()).hexdigest() == hashed_password:
            return redirect('admin')
        else:
            return render_template('login2.html')
if __name__ == '__main__':
    app.run()
````

这个例子使用了`SHA-256`哈希函数，但在实际应用中，你可能会想使用更安全的哈希函数，如`bcrypt`或`scrypt`。

Copilot Chat的回答是：

这个函数确实会造成无限循环，因为`while True:`会一直执行下去，除非有一个`break`语句或者程序被外部中断。

如果你想让这个循环在某种条件下停止，你可以添加一个条件判断和`break`语句。例如，如果你想在循环执行了10次后停止，你可以这样修改：

```python
def finite_loop():
    count = 0
    while True:
        print("This is loop number:", count)
        count += 1
        if count >= 10:
            break
```

在这个修改后的函数中，`count`变量用于跟踪循环执行了多少次。当`count`达到或超过10时，`break`语句会被执行，从而结束循环。

````python
# 定义一个有限循环函数
def finite_loop():
    # 定义一个计数变量
    count = 0
    # 开始一个无限循环
    while True:
        # 打印出当前循环的次数
        print("This is loop number:", count)
        # 计数变量加1
        count += 1
        # 当计数变量大于等于10时，跳出循环
        if count >= 10:
            break
# 调用有限循环函数
finite_loop()
````
