# 吐槽图片生成工作流

## 工作流目的
提供提示词，生成相关吐槽内容的图片。

### 使用组件

- 大模型＋图片生成
  

### 操作过程

- 1.开始节点:接收用户输入
- 2.接入大模型节点:接入大模型生成的吐槽内容。
- 3.图像生成节点:根据模版，将标题和吐槽内容放入合适的画板中
- 4.结束:输出image，画板内容

### 工作流
1.创建大模型工作流节点，设置输入输出和模型。

2.设置画板格式，包括生成内容位置和标题位置。

3.将两者组合，最后在结果处输出画板内容。
