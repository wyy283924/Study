# Coze工作流搭建之人生模拟器 

## 作流目的
该工作流的主要目的是创建一个AI智能体，模拟人生选择和路径，允许用户体验不同的人生轨迹。通过这个模拟器，用户可以在不同年龄阶段做出选择，观察这些选择如何影响他们的人生结局。

### 使用的组件
- 1.核心提示词：定义智能体的任务、功能和工作流程。
- 2.大语言模型节点：用于生成与用户选择相关的内容。
- 3.AI绘图插件：生成与用户选择相关的图片（如婴儿诞生的图片）。
- 4.工作流节点：包括开始节点、结束节点和中间的处理节点，确保整个流程的顺畅。

**（2）设定Agent模式**

创建智能体后，你会直接进入智能体编排页面。
Agent模式分为单Agent模式和多Agents模式。
单Agent模式处理复杂任务时，需要通过高质量提示词，描述AI的任务和流程，并使用提示词调动各种插件和工作流程。
多Agent模式允许你添加多个Agent，通过节点之间的分工协作，降低提示词的写作难度，但增加了设计流程和配置的复杂度。这里，我以单Agent模式为例子举例。

**(4)编写智能体核心的提示词，这是关键的步骤。**
提示词框架结构明确，要素挑选了任务框架明确，要素挑选了任务、要求、功能，人生经历，人生轨迹，工作流程，人生报告和初始等。每个要素都有清晰的指示。方便AI理解和执行。人生轨迹设计，逐步递进。从早年到晚年，每个阶段的选择，都会影响后续的经历。体现了选择的因果关系。为了增加趣味性，还增加了ImageGenerate的工作流。用于每次开始游戏时，自动生成一张婴儿诞生的图片。

### 工作流程
- 1.用户输入“人生重启按钮”后，调用ImageGenerate函数生成图片，内容为“一名中国婴儿躺在婴儿车里"，然后进入[人生轨迹]流程;
- 2.在[人生轨迹]流程中，依次选择事件，模拟人生经历，直到生命结束。完成后自动进入下一步;
- 3.按照指定的[人生报告]格式生成报告，并询问用户是否希望再次重生。如果是，请开启新的对话，重新开始。

### 项目总结
总之，高质量结构化的提示词撰写、提示词框架的选择以及提示词模式设计策略至关重要。他们不仅构建了智能体的核心，还起到了调用插件和串联整个工作流程的作用。此外提示词的效果和稳定性，不仅依赖于提示词质量本身，还依赖于大语言模型的理解能力。不同模型的能力也有着不小的差异，如果底层大模型的理解能力不足，结构过于负责或者过长的提示词，可能无法被完整的执行。此时你可能需要更换底层模型，或者使用多Agent模式，将功能分成不同的智能体。
