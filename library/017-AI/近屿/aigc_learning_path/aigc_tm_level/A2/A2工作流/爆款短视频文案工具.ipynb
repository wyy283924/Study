# 爆款短视频文案工具

## 工作流目的：抓取抖音的爆款标题，改写爆款文案

**使用的组件**:LinkRederPlugin   大模型1   大模型2
- 1.开始节点：接收用户输入。
- 2.接入抖音节点：接入抖音的视频内容
- 3.LinkRederPlugin节点：在抖音抓取爆款视频提取链接
- 4.大模型节点：大模型1负责改写标题，大模型2负责改写内容
- 5.结束节点：输出改写的结果

### 操作过程
- 1.创建工作流：在COS个人空间中新建一个工作流，命名为“dybk”。
- 2.设置开始节点：默认变量为用户输入，设置为接收用户的对话内容。
- 3.添加抖音视频节点：设置keyword为用户输入的内容，count改为2。
- 4.添加LinkRederPlugin节点：里面的url设置为list里面的link获取网址
- 5.添加大模型节点：一个负责标题 一个写文案  两个个模型里面的input设置为content，分别加入系统提示词，设定他的功能，角色 。
- 6.结束节点：引用大模型的输出结果
- 7.测试运行：运行工作流，检查生成的改写文案效果。
- 8.发布工作流：完成后发布工作流，并获取工作流ID以便在小程序中调用。
