# 可变现工作流   

## 工作流目的：
该工作流的主要目的是创建一个可接入微信小程序的系统，能够根据用户输入生成精准的吐槽，并输出一张图片。通过这个工作流，用户可以实现内容生成和流量变现。
使用的组件
•开始节点：接收用户输入。
•大模型节点：调用大模型生成吐槽内容。
•结束节点：输出生成的结果。
•图像流：生成并处理图像，最终输出为图片。


### 使用的组件
- 开始节点：接收用户输入。
- 大模型节点：调用大模型生成吐槽内容。
- 结束节点：输出生成的结果。
- 图像流：生成并处理图像，最终输出为图片。

### 流程图
- 开始节点 → 接收用户输入
- 大模型节点 → 生成吐槽内容
- 图像流 → 生成图片
- 结束节点 → 输出生成的图片

### 操作过程
- 创建工作流：在COS个人空间中新建一个工作流，命名为“demo”。
- 设置开始节点：默认变量为用户输入，设置为接收用户的对话内容。
- 添加大模型节点：选择合适的大模型（如Function Call），设置输入为用户输入的内容，并配置系统提示词。
- 设置结束节点：引用大模型的输出结果。
- 创建图像流：新建图像流，设置开始和结束节点，插入图像生成节点（如画板），并配置输入为大模型的输出。
- 连接节点：将各个节点按照执行顺序连接起来。
- 测试运行：运行工作流，检查生成的吐槽和图片效果。
- 发布工作流：完成后发布工作流，并获取工作流ID以便在小程序中调用。
