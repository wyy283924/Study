# 扣子(COZE)工作流提取文案时同步纠错

## 工作流目的
该工作流的主要目的是提取抖音文案，并在提取后同步纠错和增加标点符号，以提高文案的可读性和规范性。

### 使用的组件
- 输入网址：用于读取抖音文案的链接。
- 文本提取代码：提取文案中的标题、作者和正文。
- 大模型：用于处理文本，添加标点符号和纠正错别字。
- 飞书文档：用于上传和保存最终修改后的文案。

### 流程图
- 输入网址
- 读取文本
- 提取标题、作者和正文
- 连接大模型进行文本处理
- 上传修改后的文案到飞书文档

### 操作过程
- 输入网址：用户输入抖音文案的链接。
- 读取文本：使用代码读取链接中的文本内容。
- 提取信息：提取文案的标题、作者和正文。
- 文本处理：
- 将提取的正文传递给大模型。
- 大模型进行错别字纠正和标点符号添加。
- 上传文案：将处理后的文案上传到飞书文档，确保格式规范，便于阅读。
  
**通过这个工作流，用户能够有效地提取和处理抖音文案，提升文案的质量和可读性。**
