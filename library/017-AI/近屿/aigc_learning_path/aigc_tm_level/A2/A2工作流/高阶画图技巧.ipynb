# 高阶画图技巧
## 工作流目的
融合两个图片，并根据提示词完成替换背景

### 使用组件
- 抠图组件+融合组件+提示词组件+背景替换+风格转绘+图像生成

### 操作过程
首先进行抠图，融合。开个新分支，用于提示词生成背景。将ai生成的背景替换掉，再进行风格转绘。

### 工作流以及详细步骤
1.建立两个图像上传节点，通过融合节点完成底图和上层图的融合。

2.另起分支，使用提示词节点去完成图像生成的内容。此内容为背景，将融合后的图与背景相融合，此处使用背景替换的组件。最后使用风格转换组件，卡通风格，最后输出。
