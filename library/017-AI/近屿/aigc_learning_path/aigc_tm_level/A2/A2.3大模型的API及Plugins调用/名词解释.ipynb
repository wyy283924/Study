# 1.API:

API，全称是应用程序接口（Application Programming Interface），是一种用来让不同的软件应用程序之间相互通信和交换数据的方式。简单来说，它就像是两个人之间用来沟通的语言一样，让不同的软件或服务能够互相理解和交流。

在计算机领域中，API就是定义了一组规则和约定，让不同的软件或服务能够通过预定义的接口（比如一些函数或指令）来请求和获取数据、执行操作或者获取服务。例如，许多社交媒体平台提供API，允许开发者通过API接口访问用户数据，从而创建新的应用程序或者集成其他服务。

# 2.CMMLU（Chinese Massive Multitask Language Understanding 中文大规模多任务语言理解）

CMMLU是一个综合性的中文评估基准，专门用于评估语言模型在中文语境下的知识和推理能力。CMMLU涵盖了从基础学科到高级专业水平的67个主题，它包括：需要计算和推理的人文科学，需要知识的科学性和社会科学，需要生活常识的中国实践规则等。此外，CMMLU中的许多任务都具有中国特定的答案，可能在其他地区或语言中并不普遍适用。因此是一个完全中国化的中文评估基准。

# 3.Weight（权重）

模型weight（权重）是模型内部的一部分，用于帮助模型做出决定。你可以把它们想象成模型的大脑中的小齿轮。每个权重都是一个数字，它决定了模型在处理数据时该如何行动或做出选择。复杂的模型可以拥有百万以上的权重数据，但归根结底同样是根据这些权重来做出决定。在1.1的大模型评分表格中，Access: weight 表示该模型的测评结果并非通过网络应用，而是获得了该模型的完整权重数据，直接运行整个模型得到的。

### 实际例子

想象你有一个机器人助手，它的任务是根据你的喜好帮你选择水果。这个机器人需要考虑你对不同水果的喜好程度。我们可以把每种水果的喜好程度看作是不同的权重。根据你过去的选择，它会给每种水果分配一个权重。比如：

苹果的权重是2（表示你比较喜欢）
香蕉的权重是0.5（表示你不太喜欢）
橙子的权重是1（表示你中等喜欢）

当机器人需要帮你选择水果时，它会根据这些权重来做出决定。它会倾向于选择权重更高的水果。如果你问它今天应该吃什么水果，它可能会推荐苹果，因为苹果的权重是2，最高。

# 4.Zero-Shot(零样本)

在处理自然语言和图像的模型中，zero-shot（零样本）是指模型在没有见过任何相关训练数据的情况下完成任务的能力。这意味着模型可以处理完全新类型的问题或任务，即使在训练过程中从未见过这些具体的例子。

### 实际例子
假设我们有一个大语言模型，它在训练时学习了动物的一些常识和行为，但从未见过任何有关“海豚”行为的问题。
现在，有人问这个模型：“海豚能够表演哪些特技？”对于大模型来说，回答关于它在训练时从未见过的“海豚能够表演哪些特技？”这个问题就是一个Zero-shot任务。

具有较好Zero-shot学习能力的模型可能理解海豚是一种高度智能的海洋哺乳动物，而许多哺乳动物可以经过训练来表演特技；基于这些信息，模型可以推断出海豚可能会表演跳跃、翻滚、顶球等特技。这是因为模型能供利用它对其他动物和训练行为的理解来推断海豚的特技表演。

# 5.文本信息抽取：

文本信息抽取是指从一段文本中提取出特定的信息或数据。就像是从故事中找出关键的事情一样，文本信息抽取帮助计算机程序理解并提取出我们感兴趣的内容。

### 实际例子
假设你有一段文章讲述了动物园里不同动物的特征和习性。如果我们只想知道所有关于狮子的信息，文本信息抽取就可以帮助我们从这段文章中把关于狮子的部分提取出来，比如它们的食物、生活习性等。

在人工智能领域中，做文本信息抽取有时需要专业知识和大量数据的支持。有些任务需要从特定领域的大量文本中抽取信息，这些信息可能对于一些大型模型来说是新的，需要它们在没有特别训练数据的情况下进行零样本抽取。这就意味着，即使是先进的大模型在这些任务上的表现可能相对不足，需要额外的调整和数据支持。

# 6.JSON格式：

JSON（JavaScript Object Notation）是一种用来存储和传输数据的格式。JSON文件很容易让人和计算机读取和理解。

### JSON文件的格式

JSON文件由以下几种基本元素组成：

1. **花括号 `{}`**：用来表示一个对象（像一个盒子，里面装着成对的信息）。
2. **方括号 `[]`**：用来表示一个数组（像一个列表，里面装着很多项）。
3. **键值对 `key: value`**：用来表示具体的信息，键和值之间用冒号 `:` 分开。键是信息的名字，值是信息的内容。键和值都用双引号 `""` 括起来。

### 实际例子

假设我们要存储一个人的信息，包括姓名、年龄、地址和兴趣爱好。我们可以用JSON格式来表示这些信息：

```json
{
  "name": "小明",
  "age": 25,
  "address": {
    "street": "和平街",
    "city": "北京",
    "zipcode": "100000"
  },
  "hobbies": ["读书", "游泳", "下棋"]
}
```

**详细解释：**

1. **花括号 `{}`**：
    - 整个JSON数据包被花括号括起来，表示这是一个对象（即一组相关的信息）。
2. **键值对 `key: value`**：
    - `"name": "小明"`：键是`"name"`，值是`"小明"`。
    - `"age": 25`：键是`"age"`，值是`25`。
3. **嵌套的对象**：
    - `"address": { "street": "和平街", "city": "北京", "zipcode": "100000" }`：
        - 键是`"address"`，值是一个对象，这个对象里面又有三个键值对，分别是`"street": "和平街"`, `"city": "北京"`, 和`"zipcode": "100000"`。
4. **数组 `[]`**：
    - `"hobbies": ["读书", "游泳", "下棋"]`：
        - 键是`"hobbies"`，值是一个数组，数组里有三个字符串，分别是`"读书"`, `"游泳"`, 和`"下棋"`。

# 8.缓存、缓存命中率（2.1）

在计算机领域中，特别是在人工智能和软件开发中，我们经常需要处理大量的数据和信息。有时候，我们会使用缓存来加快数据访问速度。就像是在玩具箱里放了一些经常玩的玩具，这样你可以更快找到它们一样。

“缓存命中率”是指在使用缓存时，成功找到所需数据或信息的概率。如果你经常能从缓存中找到你需要的数据，那么缓存命中率就高。反之，如果大部分时间都需要从其他地方获取数据，缓存命中率就低。
### 实际例子
假设你有一个玩具箱，里面放了很多玩具，比如小车、积木等等。每次你需要找某个玩具时，你都能快速找到它，这时我们说你的“缓存命中率”很高。这意味着你大部分时间都能从缓存（玩具箱）中找到你想要的东西，而不用去其他地方寻找。

在文中提到的例子中，lru cache（最近最少使用缓存）是一种管理数据的方法，它确保经常使用的数据可以快速访问，同时管理缓存的上限和数据的替换，以保持高效率和成本控制。

# 9.环境变量（2.1）

在计算机里，特别是当我们编写程序或使用软件时，有时会需要一些特殊的信息，比如密码、网址或者其他重要的数据。这些信息通常被称为“环境变量”。环境变量是一种设置在计算机系统中的值，可以影响软件程序的行为。

### 实际例子
当你初始化一个OpenAI实例时，传入的api_key和base_url就是环境变量，它们告诉程序如何连接和使用OpenAI的服务。这样，你的程序就可以使用这些环境变量来初始化连接，而不需要在每次使用时手动输入这些信息。

# 10.访问令牌（Access Token）（2.1）

**访问令牌（Access Token）** 是一种用于验证身份的小“钥匙”。当你想访问某些在线服务（比如一个网站、应用程序或API）时，这个服务需要确认你是谁以及你是否有权访问它。访问令牌就是用来证明你身份的小“钥匙”。在网络上，当你登录一个网站或使用一个应用程序，它会给你一个访问令牌，这个令牌告诉服务器你已经通过了身份验证。接下来，你在一段时间内可以使用这个令牌访问你有权限的资源，而不需要每次都重新登录。

### 实际例子

假设你有一个应用程序，它需要从一个在线服务获取数据。为了安全起见，这个服务要求你先登录，然后才会给你一个访问令牌。你可以用这个令牌在接下来的请求中证明你的身份。

# 11.请求（Request），POST，请求体（Payload）（2.1）

### 什么是请求（Request）？什么是POST？

在网络上，当你访问一个网站或使用一个在线服务时，你的电脑会向服务器发送一个请求，就像你向朋友发出一个问候一样。**POST请求**是一种类型的请求，它通常用来发送数据给服务器，让服务器处理这些数据。

### 什么是请求体（Payload）？

请求体（Payload）是POST请求的一部分，里面包含了你要发送给服务器的数据。就像你写给朋友的信中的内容，信封是请求，信里面的内容就是请求体。

### 实际例子

假设你有一个在线表单，用来提交你的名字和电子邮件地址。你填写完表单后点击“提交”按钮，这个操作会发送一个POST请求给服务器，请求体里面包含了你的名字和电子邮件地址。

# 12.WebSocket（2.1）

**WebSocket**是一种在浏览器和服务器之间进行双向通信的技术。它就像一条双向的高速公路，允许数据同时在两端快速地来回传输。这和传统的HTTP请求不同，HTTP请求是单向的，每次请求都会建立一个新的连接，而WebSocket可以保持一个持续的连接，方便实时数据的传输。

### WebSocket的例子

想象你正在使用一个在线聊天应用。这个应用需要在你和你的朋友之间快速传递消息。传统的HTTP请求效率不高，因为每次发送和接收消息都需要重新建立连接。WebSocket就像开了一条永久的通道，双方可以快速互发消息。

# 13.流式传输（2.1）

**流式传输**是一种持续传输数据的方式，就像水流一样，数据不断地流动。它适用于那些需要连续发送数据的情况，比如视频直播、在线游戏、股票行情等。流式传输使得数据可以在传输的同时被接收和处理，而不需要等待所有数据传输完成后再处理。

### 流式传输的例子

想象你在观看一个在线视频，比如YouTube上的直播。当你打开视频时，视频数据是以流的形式不断地从服务器传输到你的设备上，而你可以边接收边观看，不需要等待整个视频下载完成。

# 14.上下文窗口（2.3）

在人工智能领域中，特别是在使用语言模型时，“上下文窗口”指的是模型在处理输入时考虑的文本范围或长度。就像在阅读一篇文章时，你只能看到当前页面的内容，而不能一次看到整本书一样。

举个例子，假设你在阅读一篇有关动物的故事。如果你的“上下文窗口”是整个故事的一页内容，那么你只能看到这一页中的信息。如果要了解整个故事的背景和情节，可能需要翻阅更多的页面。

在文中提到的gpt-3.5-turbo的例子中，当说“上下文窗口仅有 4096”时，意味着在处理文本时，模型只能考虑到长度为4096个字符的文本段落。如果文本太长，超过了这个限制，模型可能无法正确理解整个文本的意思，需要额外的处理或截断文本。

# 15.模型快照（2.3）

**模型快照**是指在某个特定时间点保存的AI模型的状态。你可以把它想象成拍照，每次拍照就记录下了那一刻的样子。同样，模型快照记录下了模型在那个时间点的所有参数和设置。这对于在训练模型时非常重要，因为它允许你保存当前的进度，以便之后可以继续训练或者在出现问题时恢复到之前的状态。

### 实际例子

想象你正在训练一个AI模型来识别图像中的猫和狗。训练模型可能需要很多时间和计算资源。你可能希望在训练的过程中保存一些中间状态，这样如果计算机崩溃或者需要调整训练参数，你可以从最近保存的状态继续，而不是从头开始。


# 16.SDK（2.3）

**SDK**的全称是**软件开发工具包**（Software Development Kit）。它是一个集合，里面包含了开发某个应用程序或软件所需要的工具、库、文档和示例代码。你可以把它想象成一个工具箱，里面有你需要的所有工具，帮助你更容易地开发出软件。

### 实际例子

假设你是一名应用程序开发者，你想要开发一个使用相机拍照的手机应用。你可以使用相机SDK来帮助你实现这个功能。这个SDK会包含所有你需要的东西，比如：

1. **工具和库：** 用来直接控制手机相机的代码库。
2. **文档：** 详细说明如何使用这些工具和库。
3. **示例代码：** 一些示例程序，展示如何使用这些工具来实现拍照功能。

# 17.负载均衡服务调用（2.4）

**负载均衡**是指将工作任务分配到多个服务器上，以确保每个服务器的工作量大致相同，从而提高系统的性能和可靠性。你可以把它想象成一个餐厅的服务员将顾客分配到不同的桌子上，以确保每个服务员都有相同数量的顾客要服务。

### 实际例子

假设你有一个非常受欢迎的网站，它需要处理大量的用户请求。如果所有的请求都发送到同一台服务器，这台服务器可能会很快过载，导致网站变慢甚至无法访问。为了解决这个问题，你可以使用负载均衡。当用户访问你的网站时，负载均衡器会根据服务器的当前负载情况，将请求分配给相对空闲的服务器。

# 18.Markdown格式文本和HTML格式文本

**Markdown**是一种轻量级的标记语言，用来写作和格式化文本。它使用简单的符号和语法规则，使文本更易读和编辑。Markdown特别适合用于撰写文档、博客文章、README文件等。

### HTML的主要特点：

# 19.模型微调
假设你有一把新的自行车，但是你觉得车把的高度不太合适。你只需要微调一下车把的高度，而不是重新造一辆新自行车。这样可以节省时间和精力，同时使得自行车更适合你骑行的需求。

在人工智能领域，微调就像是对一个已经训练好的模型进行小小的调整，使它更适合解决新的任务或适应新的数据。通常情况下，我们会使用一个已经在大规模数据上训练过的模型（比如预训练模型），然后通过微调来适应特定的任务或数据集。这样可以显著减少训练所需的时间和数据量，同时利用已有的模型参数和知识来加速新任务的学习过程。

# 20.可反序列化文本

**反序列化**是指将数据从一种格式转换回原来的数据结构。**可反序列化文本**是指可以被转换回原来的数据结构的文本格式。你可以把它想象成把数据保存成一种特别的格式，然后再把这种格式转换回原来的样子。
### 实际例子
假设你有一个玩具积木的模型（比如一个房子），你把这个房子拆开，并按照一定的顺序排列和记录下来。这个记录就像是序列化的文本。然后，你可以根据这个记录把积木重新拼成原来的房子，这个过程就是反序列化。


# 21.有监督

“有监督”是指一种训练模型的方法，这种方法依赖于已标注的数据。当我们说某个任务是“有监督”的时候，意味着我们为计算机提供了已经标记好的数据示例，告诉它应该如何处理和理解这些数据。你可以把它想象成一个老师在教学生，通过提供正确答案（标注的数据），帮助学生学习并改进。

### 实际例子
假如我们用有监督方式训练一个分类猫和狗照片的模型，我们使用的训练图片会带有“猫”或“狗”的标签。模型可以对照标签判断自己的结果是对还是错，得出当前结果的的正确率，并在之后的训练中提升准确率。

# 22.调参

**调参**是“调整参数”的简称。在人工智能和机器学习中，调参是指调整模型的各种设置，以便让模型在任务中表现得更好。你可以把它想象成调整一个乐器的音调，让它发出最好的声音。

想象你在烘焙蛋糕，你需要调整烤箱的温度和烤的时间，才能做出最美味的蛋糕。同样，在训练AI模型时，你需要调整模型的参数，使它能够更好地完成任务。

### 实际例子

假设我们要训练一个模型来预测房价。模型有很多参数，比如学习率、决策树的深度等。通过调整这些参数，我们可以提高模型的预测准确性。

# 23.向量

**向量**是一个有方向和大小的量。你可以把它想象成一支箭，这支箭既有长度（大小），也有指向某个方向。想象你在一个平面上画一个箭头，这个箭头从点A指向点B。箭头的长度表示它的大小，箭头的方向表示它的方向。这个箭头就是一个向量。
### 实际例子
假设你在玩积木，每个积木都有一个颜色和一个形状。你可以用一个向量来表示每个积木的颜色和形状。比如，红色的正方形可以用一个向量表示为 [1,0,0,1]，其中第一个数表示红色，后面的数表示正方形。而蓝色的圆形可以用另一个向量表示为 [0,1,1,0]，其中第二个数表示蓝色，后面的数表示圆形。

在计算机里，向量被用来表示数据和信息。比如，Word2Vec 把每个词语转换成一个向量，这样计算机就可以更好地理解词语之间的关系和意义。向量还可以帮助计算机进行数学运算和分析数据。



# 24.（文本）分块

“分块”指的是将大块的数据或文本分割成小块或片段。就像是把一块大蛋糕切成小块一样，这样做有助于更好地管理和处理数据，使得数据更容易被机器理解和处理。

### 实际例子
假如你需要让大模型阅读文章并回答你一个具体的问题。在不采用文本分块的情况下大模型会尝试理解整篇文章，并从大量的无关信息里寻找你想要的答案。这样模型生成的回答很可能不精确或者含有无关信息。
而如果大模型预先采用分块方式，只截取长文章中和你问题相关的部分，再去理解并回答，就能得到更准确的答案。

# 25.文本向量化

**文本向量化**是指将文字转换成数字形式，以便计算机可以理解和处理。你可以把它想象成把一句话翻译成一组数字。通过这种方式，计算机可以使用数学方法来分析和处理这些文字。

### 为什么需要文本向量化？

计算机擅长处理数字，但不擅长直接处理文字。为了让计算机能够理解和分析文本，我们需要把文本转换成计算机能理解的数字形式，这个过程就叫做文本向量化。

### 实际例子

假设我们有一句话：“我喜欢学习人工智能。” 我们可以按如下步骤把这句话转换成计算机可以理解的数字形式。


1. **准备词汇表：**
    - 首先，我们需要一个词汇表，把每个单词对应到一个唯一的数字。例如：
        
        ```json
        {
          "我": 1,
          "喜欢": 2,
          "学习": 3,
          "人工智能": 4
        }
        ```
        
2. **文本向量化：**
    - 通过词汇表把这句话转换成一组数字：
        
        ```csharp
        [1, 2, 3, 4]
        ```

# 26.余弦相似度

**余弦相似度**是一种用来衡量两个向量之间相似程度的方法。它通过计算两个向量之间的夹角来确定它们的相似度。余弦相似度的值介于 -1 和 1 之间，其中：

- 1 表示两个向量完全相似（方向完全相同）。
- 0 表示两个向量不相似（方向完全不同）。
- -1 表示两个向量完全相反（方向完全相反）。

你可以把向量想象成两支箭。余弦相似度就是计算这两支箭之间的角度。角度越小，箭的方向越接近，相似度就越高。


# 27.LangChain

**LangChain**是一个工具库，用于构建和管理复杂的对话和语言模型应用程序。它帮助开发者把多个语言模型（比如聊天机器人）串联在一起，使它们能够更好地协同工作，完成复杂的任务。你可以把LangChain想象成一个管道，把不同的语言模型和处理步骤连接在一起，创建一个完整的应用程序。

在构建复杂的AI应用程序时，单个模型可能无法完成所有任务。通过LangChain，开发者可以将不同的模型和处理步骤组合起来，创建一个强大的系统，能够处理更复杂的对话和任务。

# 28.OCR

**OCR**是“光学字符识别”（Optical Character Recognition）的缩写。它是一种技术，可以让计算机从图像、扫描文件或照片中识别和提取文字内容。你可以把它想象成一种“读图”的能力，计算机可以看到图片上的文字，并将它们转换成可编辑的文本。

OCR技术非常有用，因为它可以帮助我们将纸质文档、书籍、手写笔记等内容转换成电子文本。这不仅方便存储和检索，还可以对这些文本进行编辑和分析。


# 29.大模型幻觉

**大模型幻觉**（AI Hallucination）是指AI模型，尤其是大型语言模型，生成的内容看起来合理但实际上是不准确或虚假的。你可以把它想象成AI模型在“瞎编”或“想象”一些信息，这些信息虽然语法上正确，但在事实层面上是错误的或不存在的。



### 实际例子

假设你在使用一个AI聊天机器人，并向它询问一些信息。如果这个AI模型没有关于某个特定问题的确切数据，它可能会生成一个听起来合理但实际上不正确的回答。

你问AI模型：“地球上最高的山峰是什么？”

AI模型可能正确地回答：“珠穆朗玛峰是地球上最高的山峰，海拔约8848米。”

但是，如果你问一个更具体的问题，比如：“珠穆朗玛峰旁边的第二高的山叫什么名字？”

AI模型可能会回答：“珠穆朗玛峰旁边的第二高的山是长城山，海拔8500米。”

### 为什么会发生大模型幻觉？

大模型通过学习大量的文本数据来生成内容，但它们并不具备真正的理解能力。它们生成内容的依据是模式和概率，而不是对事实的深刻理解。因此，有时候模型会生成一些看起来正确但实际上不正确的信息。
